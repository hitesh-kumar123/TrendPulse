# Sentiment Analysis on Movie Reviews

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

### 1. Load Dataset
Loading the standard 50,000 IMDb movie reviews dataset.

```python
df = pd.read_csv('../data/reviews_50k.csv')
df.head()
```

### 2. Feature Extraction (TF-IDF Vectorizer)
We use `TfidfVectorizer` to convert text data into numerical features. We will save this exact vectorizer mapping as `tranform.pkl` so the app processes live user input with the identical vocabulary.

```python
X = df['review']
y = df['sentiment']

# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

# Fit and transform the data
X_vec = vectorizer.fit_transform(X)

print('Shape of transformed data:', X_vec.shape)
```

### 3. Train the Model
We use a `MultinomialNB` classifier which works incredibly well for text classification tasks.

```python
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

clf = MultinomialNB()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nClassification Report:\n', classification_report(y_test, y_pred))
```

### 4. Export Models (`.pkl`)
Finally, export the trained classifier as `nlp_model.pkl` and the vectorizer as `tranform.pkl` into the `models` folder.

```python
# Save the model
pickle.dump(clf, open('../models/nlp_model.pkl', 'wb'))

# Save the vectorizer (NOTE: naming it tranform.pkl for app.py compatibility)
pickle.dump(vectorizer, open('../models/tranform.pkl', 'wb'))

print('Models successfully exported to ../models/ directory!')
```
