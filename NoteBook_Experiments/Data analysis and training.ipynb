### Word Stemming (Simplifying Words)
We use `PorterStemmer` to cut down words to their base root. 
For example, the words **'action'**, **'actions'**, and **'actionable'** all mean the exact same thing. But a computer might treat them as 3 completely different words.
Stemming cuts them all down to just **'action'**. This helps the model realize they mean the same thing and improves our movie recommendations!


```python
from nltk.stem.porter import PorterStemmer   
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    
    for i in text.split():
        T.append(ps.stem(i))
    
    return " ".join(T)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

### Vectorization: Converting Text to Numbers

```python
# Stick with this — it's the right choice
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=15000,
    stop_words='english',
    ngram_range=(1, 2),   # captures "Tom Hanks" as one unit
    min_df=2,
    sublinear_tf=True     # extra dampening for repeated terms
)
```

```python
vectors = tfidf.fit_transform(new_df['tags'])
```

```python
vectors[0]
```

```python
vectors.shape
```

```python
len(tfidf.get_feature_names_out())
```

### Finding the Most Similar Movies (Cosine Similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Insidious')
```
